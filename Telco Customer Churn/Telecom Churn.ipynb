## 3. Limpeza dos dados

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True).copy()
```

```python
# Verificando o dataset com o index ajustado após a exclusão das linhas com missing
df
```

```python
# O costumerID atrapalha o onehot enconding por ser um object e por isso vamos elimina-lo.

df.drop(columns= 'customerID', axis=1, inplace=True)
```

## 4. Pré-preocessamento de dados

```python
# Transformar as features em numéricas

df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

df = pd.get_dummies(df)

df.head()
```

### Gráfico de correlação das features com churn

```python
plt.figure(figsize=(20,8))
df.corr()["Churn"].sort_values(ascending = False).plot(kind='bar')
```

```python
# Separando as variaves de entrada e de saída

X = df.drop(columns='Churn').values
Y = df['Churn'].values
```

```python
# Separando em treino e teste

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)
```

## 5. Criando as máquinas preditivas


```python
# Vamos testar os modelos com crossvalscore

modelo_1 = KNeighborsClassifier()

resultado_1 = cross_val_score(modelo_1, X_train, y_train)

resultado_1.mean()*100
```

```python

modelo_2 =  DecisionTreeClassifier()

resultado_2 = cross_val_score(modelo_2, X_train, y_train)

resultado_2.mean()*100
```

```python

modelo_3 =  XGBClassifier()

resultado_3 = cross_val_score(modelo_3, X_train, y_train)

resultado_3.mean()*100
```

```python
#Tunnando os modelos 

param_space = {'n_neighbors': Integer(1, 10),
               'p': Integer(1, 2),
               'weights': Categorical(['uniform', 'distance'])}


search_1 = BayesSearchCV(modelo_1, param_space, n_iter=10, cv=5, random_state=42)

search_1.fit(X_train, y_train)

best_1 = search_1.best_params_

modelo_01 = KNeighborsClassifier(**best_1) #criando o modelo tunnado 1

print(search_1.best_score_)
```

```python
param_space_2 = {'max_depth': Integer(1, 10),
               'min_samples_split': Integer(2, 10),
               'min_samples_leaf': Integer(1, 5),
               'criterion': Categorical(['gini', 'entropy'])}


search_2 = BayesSearchCV(modelo_2, param_space_2, n_iter=10, cv=5, random_state=42)

search_2.fit(X_train, y_train)

best_2 = search_2.best_params_

modelo_02 = DecisionTreeClassifier(**best_2) #criando o modelo tunnado 2

print(search_2.best_score_)

```

```python
param_space_3 = {'learning_rate': Real(0.01, 1.0, prior='log-uniform'),
               'max_depth': Integer(1, 10),
               'subsample': Real(0.5, 1.0, prior='uniform'),
               'colsample_bytree': Real(0.5, 1.0, prior='uniform'),
               'gamma': Real(0, 5, prior='uniform'),
               'reg_alpha': Real(0, 1, prior='uniform'),
               'reg_lambda': Real(0, 1, prior='uniform')}


search_3 = BayesSearchCV(modelo_3, param_space_3, n_iter=10, cv=5, random_state=42)

search_3.fit(X_train, y_train)

best_3 = search_3.best_params_

modelo_03 = XGBClassifier(**best_3) #criando o modelo tunnado 3

print(search_3.best_score_)
```

```python
# Agora usando os dados de teste e utilizando as métricas do classification report e confusion matrix

# primeiro temos que passar os hiperparametros tunnados para cada modelo

modelos = [modelo_01, modelo_02, modelo_03]

for i in modelos:
    i.fit(X_train, y_train)
    p = i.predict(X_test)
    print(classification_report(y_test, p))
    print(confusion_matrix(y_test, p))
```

Notamos que o modelo se sai muito bem quando prevê o não-churn, mas não consegue prever quando ocorre de fato o churn, principalmente o recall do melhor modelo (XGBoost), indicando assim que acertou apenas 50% dos clientes que cancelaram a assinatura.

Como observamos no ínicio, as classes estão desbalanceadas: temos os dados de 5174 clientes que não deram churn contra 1869 que deram. Então aplicaremos a técnica de balanceamento SMOTE

## Balanceando os dados e criando uma nova máquina preditiva.

```python
from imblearn.over_sampling import SMOTE

# Cria o balanceador SMOTE
smote_bal = SMOTE()

# Aplica o balanceador
X_res, y_res = smote_bal.fit_resample(X, Y)
```

```python
# Dar o slpit novamente

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size = 0.20, random_state = 0)
```

```python
modelor_1 = KNeighborsClassifier()

resultador_1 = cross_val_score(modelor_1, X_train, y_train)

print(resultador_1.mean()*100)



modelor_2 =  DecisionTreeClassifier()

resultador_2 = cross_val_score(modelor_2, X_train, y_train)

print(resultador_2.mean()*100)


modelor_3 =  XGBClassifier()

resultador_3 = cross_val_score(modelor_3, X_train, y_train)

print(resultador_3.mean()*100)
```

```python
# Vamos usar o XGBoost que teve o melhor resultado

param_space_3r = {'learning_rate': Real(0.01, 1.0, prior='log-uniform'),
               'max_depth': Integer(1, 10),
               'subsample': Real(0.5, 1.0, prior='uniform'),
               'colsample_bytree': Real(0.5, 1.0, prior='uniform'),
               'gamma': Real(0, 5, prior='uniform'),
               'reg_alpha': Real(0, 1, prior='uniform'),
               'reg_lambda': Real(0, 1, prior='uniform')}


search_3r = BayesSearchCV(modelor_3, param_space_3r, n_iter=10, cv=5, random_state=42)

search_3r.fit(X_train, y_train)

best_3r = search_3.best_params_

modelo_03r = XGBClassifier(**best_3r) #criando o modelo tunnado 3

print(search_3.best_score_)
```

```python
# Agora finalemnte vamos verificar as métricas que estavam baixas anteriormente

modelo_03r.fit(X_train, y_train)

pr = modelo_03r.predict(X_test)

print(classification_report(y_test, pr))
print(confusion_matrix(y_test, pr))
```

Então temos um modelo cuja acurácia de 86% realmente acerta para as duas classes, de churn e de não churn.
