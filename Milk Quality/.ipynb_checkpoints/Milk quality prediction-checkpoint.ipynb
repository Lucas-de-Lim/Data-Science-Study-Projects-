## 3. Pré-preocessamento de dados

```python
# Tornando numérico os dados categóricos de saída
label_encoder = preprocessing.LabelEncoder()

df['Grade'] = label_encoder.fit_transform(df['Grade']) 
```

```python
# Separando os dados 

X = df.iloc[:, :-1].values
Y = df.iloc[:, -1] .values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

## 4. Criando a máquina preditiva

### Regressão logística

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=7)
```

```python
modelo_1 = LogisticRegression()

resultado_1 = cross_val_score(modelo_1,X_train, Y_train, cv=kfold)

print(resultado_1.mean())
```

### RandomForest

```python
modelo_2 = RandomForestClassifier()

resultado_2 = cross_val_score(modelo_2,X_train, Y_train, cv=kfold)

print(resultado_2.mean())
```

### XGboost

```python
modelo_3 = XGBClassifier()

resultado_3 = cross_val_score(modelo_3, X_train, Y_train, cv=kfold)

print(resultado_3.mean())
```

```python
# Vendo a acurácia nos treinos de teste

modelos = [modelo_1, modelo_2, modelo_3]

for i in modelos:
    i.fit(X_train, Y_train)
    p = i.predict(X_test)
    print(classification_report(Y_test, p))
    print(confusion_matrix(Y_test, p))
```

```python
# Resultamos então numa altissima performance do RandomForest e XGBoost para a classificação desse dataset, chegando a acertar
# tudo dos dados de treino. Dispensando até mesmo o tunning de hiperparâmetros
```
