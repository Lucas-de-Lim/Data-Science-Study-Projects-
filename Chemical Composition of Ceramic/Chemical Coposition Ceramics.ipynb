```python
# Bibliotecas para manipulção de dados
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

# Bibliotecas para preprocessamento de dados 
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# Bibliotecas utilizadas na Construção de Máquinas Preditivas
from sklearn.cluster import KMeans


# Bibliotecas utilizadas na Avaliação das Máquinas
from scipy.spatial.distance import cdist, pdist
from sklearn.metrics import silhouette_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Chemical Composion of Ceramic.csv')
```

```python
df_x = df.drop(columns=['Ceramic Name','Part']).values
```

```python
df_x
```

```python
pca = PCA(n_components = 2).fit_transform(df_x)
```

```python
# Determinando um range do Hyperparâmetro  "K"  do Kmeans
k_range = range(1,12)
k_range

# Aplicando o modelo K-Means para cada valor de K (esta célula pode levar bastante tempo para ser executada)
k_means_var = [KMeans(n_clusters = k).fit(pca) for k in k_range]### Curva de Elbow
```

### Curva de Elbow

```python
# Ajustando o centróide do cluster para cada modelo
centroids = [X.cluster_centers_ for X in k_means_var]

# Calculando a distância euclidiana de cada ponto de dado para o centróide
k_euclid = [cdist(pca, cent, 'euclidean') for cent in centroids]
dist = [np.min(ke, axis = 1) for ke in k_euclid]

# Soma dos quadrados das distâncias dentro do cluster
soma_quadrados_intra_cluster = [sum(d**2) for d in dist]

# Soma total dos quadrados
soma_total = sum(pdist(pca)**2)/pca.shape[0]

# Soma dos quadrados entre clusters
soma_quadrados_inter_cluster = soma_total - soma_quadrados_intra_cluster

# Curva de Elbow
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range, soma_quadrados_inter_cluster/soma_total * 100, 'b*-')
ax.set_ylim((0,100))
plt.grid(True)
plt.xlabel('N° de Clusters')
plt.ylabel('% de Variância Explicada')
plt.title('Variância Explicada para cada Valor de K')
```

### Escolhendo um Valor de K inicial (igual a 2) para avaliação da Máquina Preditiva

```python
# Criando um modelo com K = 2
modelo_v1 = KMeans(n_clusters = 2)
modelo_v1.fit(pca)
```

```python
df['cluster'] = modelo_v1.labels_
```

```python
df['cluster'].value_counts()
```

```python
sns.countplot( data=df, x = 'Part', hue='cluster')
plt.show()
```

Separando os dados em 2 clusters através do kmeans podemos ver que ele coincide com a feature 'Part' que distingue entre dois tipos a cerâmica.
